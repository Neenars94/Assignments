# Data Preprocessing

```python
import string
```

```python
string.punctuation
```

```python
# text cleaning- removing punctuations
def remove_punctuation(text):
  punctuation_free="".join([i for i in text if i not in string.punctuation])
  return punctuation_free
```

```python
# Tokenization
import nltk
nltk.download('punkt_tab')
def tokenization(text):
    words = nltk.word_tokenize(text)
    return words
```

```python
#stop words removal
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output
```

```python
#stemming/lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
wordnet_lemmatizer = WordNetLemmatizer()
def lemmatizer(text):
  lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
  return lemm_text
```

```python
import re
```

```python
def remove_noisy_token(tweet):
    return re.sub(pattern=r'@[a-zA-Z0-9 ]+|#[a-zA-Z0-9 ]+|\w+:\/{2}[\d\w-]+(\.[\d\w-]+)(?:(?:\/[^\s/]))|\W+|\d+|<("[^"]"|\'[^\']\'|[^\'">])>|_+|[^\u0000-\u007f]+',string=tweet,repl=" ")
```

```python
def remove_http_link(tweet):
    return re.sub(pattern=r'https?://[^\s\n\r]+',string=tweet,repl=" ")
```

```python
# creating a function for preprocessing
def preprocess(df_col):
  corpus = []
  for item in df_col:
    new_item = remove_punctuation(item)
    new_item = new_item.lower()
    new_item = tokenization(new_item)
    new_item = remove_stopwords(new_item)
    new_item = lemmatizer(new_item)
    corpus.append(' '.join(str(x) for x in new_item))
  return corpus
```

# Feature Extraction

```python
def vectorizer_text(corpus):#(corpus,n)
  bow_model = CountVectorizer()#max_features=n
  dense_vec = bow_model.fit_transform(corpus).todense() #to convert to vector other wise we will get in array format
  bow_df = pd.DataFrame(dense_vec)
  bow_df.columns = sorted(bow_model.vocabulary_)
  return bow_df
```

```python
# 1. Bag of Words
vectorizer = CountVectorizer()
bow_matrix = vectorizer.fit_transform(corpus)
print("Bag of Words Matrix:\n", bow_matrix.toarray())
print("Vocabulary:", vectorizer.vocabulary_)
```

```python
# 2. TF-IDF
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)
print("\nTF-IDF Matrix:\n", tfidf_matrix.toarray())
print("Vocabulary:", tfidf_vectorizer.vocabulary_)
```

```python
# 3. Word2Vec
from gensim.models import Word2Vec
tokenized_corpus = [nltk.word_tokenize(sentence) for sentence in corpus]
model_w2v = Word2Vec(tokenized_corpus, min_count=1, vector_size=100) #vector_size is the embedding dimension
print("\nWord2Vec Model Vocabulary:", model_w2v.wv.key_to_index)
```

```python
# 4. Average Word2Vec
def average_word_vectors(words, model, vocabulary, num_features):
    feature_vector = np.zeros((num_features,), dtype="float64")
    nwords = 0.
    for word in words:
        if word in vocabulary:
            nwords = nwords + 1.
            feature_vector = np.add(feature_vector, model.wv[word])
    if nwords:
        feature_vector = np.divide(feature_vector, nwords)
    return feature_vector

sentence_vectors_w2v = []
for sentence in tokenized_corpus:
    sentence_vectors_w2v.append(average_word_vectors(sentence, model_w2v, model_w2v.wv.key_to_index, 100))

print("\nAverage Word2Vec Sentence Vectors:\n", np.array(sentence_vectors_w2v))
```

```python
# 5. TF-IDF weighted Word2Vec
def tfidf_weighted_word_vectors(words, tfidf_matrix, tfidf_vectorizer, model, vocabulary, num_features):
    feature_vector = np.zeros((num_features,), dtype="float64")
    tfidf_sum = 0
    for word in words:
        if word in vocabulary:
            word_index = tfidf_vectorizer.vocabulary_[word]
            tfidf_value = tfidf_matrix[corpus.index(" ".join(words)), word_index]
            feature_vector = np.add(feature_vector, np.multiply(model.wv[word], tfidf_value))
            tfidf_sum += tfidf_value
    if tfidf_sum:
        feature_vector = np.divide(feature_vector, tfidf_sum)
    return feature_vector
```
